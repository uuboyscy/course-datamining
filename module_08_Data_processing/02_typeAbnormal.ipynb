```python
columns_dtypes = {
    "DepartmentID": str,
    "CreatedDatetime": "datetime64[ns]",
}

df = pd.read_csv("dtype_sample.csv", dtype=columns_dtypes)
```

```python
df
```

```python
np.where(
    df["Experience"].isna(),
    0,
    df["Experience"],
)
```

```python
df["Experience"] = np.where(
    df["Experience"].isna(),
    0,
    df["Experience"],
)

df
```

```python
df["OnBoard"] = np.where(
    df["OnBoard"].str.strip() == "V",
    True,
    False,
)

df
```

```python
columns_dtypes = {
    "Name": str,
    "Experience": float,
    "Age": int,
    "Department": str,
    "DepartmentID": str,
    "Remote": bool,
    "Intern": bool,
}

df = pd.read_csv(
    "dtype_sample.csv",
    dtype=columns_dtypes,
    parse_dates=["CreatedDatetime", "CreatedDatetimeTZ"],
    date_parser=lambda x: pd.to_datetime(x, errors="coerce"),
)

df
```
